```python
import base64
from PIL import Image
import google.cloud.automl as automl
import tempfile
import os
import json

def predict(data):
    
    if data.method == 'OPTIONS':
        headers = {
            'Access-Control-Allow-Origin': '*',
            'Access-Control-Allow-Methods': 'POST',
            'Access-Control-Allow-Headers': 'Content-Type',
            'Access-Control-Max-Age': '3600'
        }
        
        return ('', 204, headers)

    encoded_string = bytes(data.get_json()['imageBytes'], 'utf-8')

    project_id = "ford-gas-gcp"
    model_id = "IOD7088010504551006208"
    prediction_client = automl.PredictionServiceClient()
    model_full_id = prediction_client.model_path(project_id, "us-central1", model_id)

    imgdata = base64.b64decode(encoded_string)

    _, temp_local_filename = tempfile.mkstemp()
    temp_local_filename_ext = temp_local_filename + '.jpg'

    with open(temp_local_filename_ext, 'wb') as f:
        f.write(imgdata)

    with Image.open(temp_local_filename_ext) as im:
        im_resized = im.resize((1024, 1024))
        im_resized.save(temp_local_filename_ext)

    with open(temp_local_filename_ext, "rb") as content_file:
        content3 = content_file.read()

    image = automl.types.Image(image_bytes=content3)
    payload = automl.types.ExamplePayload(image=image)

    params = {"score_threshold": "0.5"}
    response = prediction_client.predict(model_full_id, payload, params)

    print("Prediction results:")
    resp = []
    for result in response.payload:
        print("Predicted class name: {}".format(result.display_name))
        print(
            "Predicted class score: {}".format(
                result.image_object_detection.score
            )
        )
        bounding_box = result.image_object_detection.bounding_box
        resized_vertices = []
        print("Normalized Vertices:")
        vertices = []
        for vertex in bounding_box.normalized_vertices:
            print("\tX: {}, Y: {}".format(vertex.x, vertex.y))
            vertices.append(
                {
                    'x': vertex.x,
                    'y': vertex.y
                }
            )
        resp.append(
            {
                'display_name': result.display_name,
                'class_score': result.image_object_detection.score,
                'vertices': vertices
            }
        )

    os.remove(temp_local_filename)
    os.remove(temp_local_filename_ext)
    
    headers = {
        'Access-Control-Allow-Origin': '*'
    }

    return (json.dumps(resp), 200, headers)
```
